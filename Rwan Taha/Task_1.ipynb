```python
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Feature and target data
X = np.array([[1], [2], [3], [4], [5], [6]])
y = np.array([30, 35, 40, 45, 50, 60])

# Train model
model = LinearRegression()
model.fit(X, y)

# Predict for 7 years
X_test = np.array([[7]])
y_pred = model.predict(X_test)
print(f"Predicted salary for 7 years: {y_pred[0]:.2f}k")

# Plot points and regression line
plt.scatter(X, y)
plt.plot(X, model.predict(X))
plt.xlabel('Years of Experience')
plt.ylabel('Salary (k)')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
import numpy as np
import matplotlib.pyplot as plt

# Data: study hours (X) and pass/fail (y: 1=pass, 0=fail)
X = np.array([[1], [2], [3], [4], [5], [6], [7]])
y = np.array([0, 0, 0, 1, 1, 1, 1])

# Create and train the model
model = LogisticRegression()
model.fit(X, y)

# Predict for 4.5 hours
X_test = np.array([[4.5]])
y_pred = model.predict(X_test)
print(f"Prediction for 4.5 hours: {'Pass' if y_pred[0] == 1 else 'Fail'}")


X_plot = np.linspace(1, 7, 100).reshape(-1, 1)
y_prob = model.predict_proba(X_plot)[:, 1]

plt.scatter(X, y, label='Data')
plt.plot(X_plot, y_prob, label='Pass Probability')
plt.xlabel('Study Hours')
plt.ylabel('Pass Probability')
plt.title('Logistic Regression - Pass Prediction')
plt.legend()
plt.show()
```
